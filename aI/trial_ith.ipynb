```python
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
names = np.array(['wildtype',
        'G248A-G249A', 
        'G248A-G249P',
        'G248P-G249A',
        'G248P-G249P',
        'constrained',
        'V247A-G248A-G249A-L250A',
        'V247A-G248A-G249A',
        'G249W',
        'G249V',
        'G249L', 'G249I',
        'G248A_G249V', 'G248A_G249L', 'G248A_G249I'])
```

```python
colors = np.array([
    'dimgrey', 
    'yellowgreen', 'darkseagreen', 'olive', 'green',
    'black', 
    'royalblue', 'dodgerblue',
    'crimson', 'darkorange',
    'wheat', 'gold', 
    'burlywood', 'tan', 'khaki'
])
```

```python
bendings = [
    np.loadtxt(f'../7_ith_kink/4_kink_check_mutants/2_bending_angle/{i}.xvg',comments=['@','#'])[:,1]
for i in range(len(names))]
```

```python
bins = np.arange(0,182,2)
xb = bins[:-1]+1
```

```python
hbendings = np.array([np.histogram(d, bins=bins, density=True)[0] for d in bendings])
```

```python
names[[0,5,8,9]]
```

```python
names[[0,5,8,9]], names[-4:] = names[-4:], names[[0,5,8,9]]
```

```python
colors[[0,5,8,9]], colors[-4:] = colors[-4:], colors[[0,5,8,9]]
```

```python
hbendings[[0,5,8,9]], hbendings[-4:] = hbendings[-4:], hbendings[[0,5,8,9]]
```

```python
fig,ax=plt.subplots(1,figsize=(16,6))
for a,h in enumerate(hbendings):
    ax.plot(xb, h, color=colors[a], lw=3)
ax.set_xlim(60,182)
ax.set_xticks([60,100,140,180])
ax.set_ylim(-0.001,)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.spines[['top','left','right']].set_visible(False)
ax.spines['bottom'].set_linewidth(2)
ax.tick_params(labelsize=12, length=9, width=2, labelleft=False, left=False)
ax.tick_params(which='minor', length=6, width=2)
#
plt.savefig('saved_plots/trial_kiinks.pdf', transparent=True)
plt.show()
```

```python
for a,h in enumerate(hbendings): plt.plot(xb,h,lw=4, color=colors[a], label=names[a])
plt.legend()
plt.savefig('saved_plots/trial_labels.pdf')
plt.show()
```
