```python
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
names1 = ['apo', 'bound', 'allo-apo', 'allo-bound', 'allo-both']
names2 = ['g2489a', 'g249v', 'g249w', 'og249w']
```

```python
bins = np.arange(0.4,2,0.01)
xb = bins[1:] - (bins[1]-bins[0])/2
zeros = np.zeros(xb.shape)
```

```python
wt = [np.loadtxt(f'../3_channel_2_state/1_S83_S102_ca_distance/channel2_S83_S102_p450_4jx1_{n}.xvg',
                comments=['@','#'])[:,1]
     for n in names1]
```

```python
hwt = [np.histogram(d, bins=bins, density=True)[0] for d in wt]
```

```python
for h in hwt[:5]: plt.plot(xb,h)
plt.xlim(0.4,1.8)
```

```python
closed = [np.loadtxt(f'../3_channel_2_state/1_S83_S102_ca_distance/channel2_S83_S102_p450_{n}_apo.xvg',
                    comments=['@','#'])[:,1]
         for n in names2]
```

```python
hclosed = [np.histogram(d, bins=bins, density=True)[0] for d in closed]
```

```python
for h in hclosed: plt.plot(xb,h)
plt.xlim(0.4,1.8)
```

```python
opened = [np.loadtxt(f'../3_channel_2_state/1_S83_S102_ca_distance/channel2_S83_S102_p450_l166a_{n}.xvg',
                    comments=['@','#'])[:,1]
         for n in names1[:2]]
```

```python
hopened = [np.histogram(d, bins=bins, density=True)[0] for d in opened]
```

```python
for h in hopened: plt.plot(xb,h)
plt.xlim(0.4,1.8)
```

```python
decoupled = [np.loadtxt(f'../3_channel_2_state/1_S83_S102_ca_distance/channel2_S83_S102_p450_f163a_{n}.xvg',
                    comments=['@','#'])[:,1]
         for n in names1]
```

```python
hdecoupled = [np.histogram(d, bins=bins, density=True)[0] for d in decoupled]
```

```python
for h in hdecoupled: plt.plot(xb,h)
plt.xlim(0.4,1.8)
```

```python
decoupled2 = [np.loadtxt(f'../3_channel_2_state/1_S83_S102_ca_distance/channel2_S83_S102_p450_multi_{n}.xvg',
                    comments=['@','#'])[:,1]
         for n in names1]
```

```python
curve = hwt[0]
oname='ch2_wt_sf'
fig,ax=plt.subplots(1, figsize=(6,4))
ax.fill_between(xb, zeros, curve, color='crimson', alpha=0.3)
ax.plot(xb, curve, color='crimson', lw=3)
ax.set_xlim(0.4,1.8)
ax.set_xticks([0.4,1.1,1.8])
ax.set_ylim(0,)
ax.vlines(0.8, 0, ax.get_ylim()[1]/2, linestyle='--', lw=3, color='black')
ax.spines[['top','left','right']].set_visible(False)
ax.spines['bottom'].set_linewidth(2)
ax.tick_params(labelleft=False, left=False, labelsize=20, length=9, width=2)
ax.xaxis.set_minor_locator(AutoMinorLocator(3))
ax.tick_params(which='minor', length=5, width=2)
#
plt.savefig(f'saved_plots/{oname}.pdf', transparent=True)
plt.show()
```
